```python
correct_category = {
    'Shoes': 'Fashion',
    'T-Shirt': 'Fashion',
    'Jeans': 'Fashion',
    'Laptop': 'Electronics',
    'Mobile': 'Electronics',
    'Headphones': 'Electronics',
    'Refrigerator': 'Home Appliances',
    'Washing Machine': 'Home Appliances',
    'Sofa': 'Furniture',
    'Table': 'Furniture',
    'Watch': 'Electronics'
}
```

```python
data['Category'] = data['Product_Name'].map(correct_category)
```

```python
data.head()
```

```python
category_sales = data.groupby('Category')['Sales'].sum()
category_sales
```

```python
city_sales = data.groupby('City')['Sales'].sum()
city_sales
```

```python
monthly_sales = data.groupby(data['Order_Date'].dt.month)['Sales'].sum()
```

```python
monthly_sales
```

```python
top_products = data.groupby('Product_Name')['Quantity'].sum()
top_products
```

```python
Channel_sales = data.groupby('Channel')['Sales'].sum()
Channel_sales
```
